```python
# Конкатенация и добавление в конец

def make_df(cols, ind):
    """Make dataframe"""
    data = {c:[str(c)+str(i) for i in ind]for c in cols}
    return pd.DataFrame(data, ind)
make_df('ABC', range(3))
```

```python
# Фильтрация
def filter_func(x):
    return x['data2'].std() > 4
print(df)
print(df.groupby('key').std())
print(df.groupby('key').filter(filter_func))
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y) * np.cos(x)
```

```python
x_iris = iris.drop('species', axis =1)
x_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
# Пример
rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2*x - 1 + rng.randn(50)
plt.scatter(x, y)
```

```python
# 1. Выбор класса модели
from sklearn.linear_model import LinearRegression
# 2. Выбор гиепрпараметров модели
model = LinearRegression(fit_intercept = True)
model
```

```python
# 3. Формирование из данных матриц признаков и целевого вектора
X = x[:, np.newaxis]
X.shape
```

```python
# 4. Применение модели к данным
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# 5. Предсказание меток для новых данных
xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
plt.scatter(x, y)
plt.plot(xfit, yfit)
```

```python
from sklearn.cross_validation import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(x_iris, y_iris, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```
